# Support Vector Machine (SVM)

## Key Concepts in SVM

* A supervised machine learning algorithm used primarily for **classification** tasks
* It can also be applied to regression problems (via Support Vector Regression).

SVM works by ***finding the hyperplane*** that best separates different classes in a high-dimensional feature space. In SVMs, the main idea is the margin.



---



*Binary classifier*

SVM is inherently a binary classifier, but it can be extended to handle multiclass classification using two common strategies:

* One-vs-One : For K classes, build K(K−1) / 2  classifiers, each trained on a pair of classes.

* One-vs-Rest: For K classes, build K classifiers, each distinguishing one class from the rest.



---



 Hyperplane

* A hyperplane is ***a decision boundary*** that separates data points belonging to different classes.

* In a 2D space, it’s simply a line; in 3D, it’s a plane, and in higher dimensions, it’s a hyperplane.

Support Vectors

* Support vectors are the ***data points closest to the hyperplane***. These points are crucial because they determine the position of the hyperplane.

* Only these points affect the decision boundary.

Margin

* Margin is the ***distance between the hyperplane and the nearest support vectors*** of any class.

* SVM aims to maximize the margin between the classes. A larger margin usually results in better generalization.

Hard margin vs Soft margin

The main difference between hard margin and soft margin is how well the data can be separated:

* ***Hard margin*** is used when data is ***linearly separable***, meaning it can be perfectly separated into two classes without misclassification.

* ***Soft margin*** is used when data is ***not linearly separable*** or when ***some misclassification*** is acceptable

#### ***Constraints for Classification***

For the hyperplane to correctly classify all points:

> W.X + b ≥ 1  for Yi = +1  

> W.X + b ≤ -1  for Yi = -1

These constraints can be combined into a single constraint for all points:

> Yi (W.Xi + b) - 1 ≥ 0 where Yi = +1,-1 for all i=1,2,3..L

#### Introducing ***Slack Variables***

In many real-world scenarios, the data is not perfectly separable (due to noise, overlap, or outliers). To handle this, we introduce a ***soft margin***, allowing some misclassifications in the training set.


To allow for some misclassification, we introduce slack variables ξi ≥ 0 for each data point i. These slack variables represent the distance by which a point violates the margin constraint.

* If ξi = 0, the point is correctly classified and on the correct side of the margin.

* If ξi > 0, the point is misclassified or lies inside the margin.


#### New constraints for classification

The constraint for correct classification is modified to:

> yi (w⋅xi +b) ≥ 1 − ξi , ∀i

This means:

* If ξi = 0, the point is correctly classified and at least at a distance of 1 from the margin.

* If ξi>0, the point lies inside the margin or is misclassified.

## **Kernel Trick in SVM**

It is used to efficiently handle non-linearly separable data by implicitly mapping it into a higher-dimensional space, where it becomes linearly separable.

This technique avoids the need to explicitly calculate the transformation into the higher-dimensional space, making it computationally efficient.

Recall:
[What if data points is not linearly separable?](#what-if-data-points-is-not-linearly-separable)

#### Mapping Data to a Higher-Dimensional Space

For example, if we have 2-dimensional data points
x=(x1, x2), we might map them to a higher-dimensional space where they become 3-dimensional

(e.g., x → ϕ(x) = (x1^2, 2.x1.x2, x2^2)).

The decision function in the new space would then be:

> f(x)=sign(w⋅ϕ(x)+b)


mapping data into a higher-dimensional space sounds appealing, but doing so directly can be ***computationally expensive***

### Kernel Function
A kernel function K(xi, xj) computes the dot product in the transformed feature space without ever explicitly calculating the transformation ϕ(x).

The general form of a kernel is:

> K(xi,xj) = ϕ(xi)⋅ϕ(xj)

### Commonly Used Kernels

1.   Linear Kernel : K(xi,xj) = xi⋅xj

2.   Polynomial Kernel  : K(xi,xj) = (xi⋅xj + c)^d

where c is a constant and d is the degree of the polynomial.


3.   Gaussian Radial Basis Function (RBF) Kernel : K(xi,xj) = exp(− ∥xi−xj∥^2 / 2σ^2)

4.   Sigmoid Kernel  : K(xi,xj) = tanh(αxi⋅xj+c)

## Limitations of SVM


* **Scalability:** SVMs can be slow to train on large datasets, as the complexity is generally O(n^2) or O(n^3), depending on the implementation.

* **Choice of Kernel:** Selecting the right kernel and tuning its parameters can be difficult and time-consuming.

* **Interpretability:** SVMs can be harder to interpret compared to other models like decision trees.

## Code

```python
import numpy as np

class SVM:
    def __init__(self, learning_rate=0.001, lambda_param=0.01, n_iters=1000):
        self.learning_rate = learning_rate
        self.lambda_param = lambda_param
        self.n_iters = n_iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        y_ = np.where(y <= 0, -1, 1)  # Convert labels to -1 and 1

        # Initialize weights and bias
        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.n_iters):
            for idx, x_i in enumerate(X):
                condition = y_[idx] * (np.dot(x_i, self.w) + self.b) >= 1
                if condition:
                    # Gradient for correctly classified points
                    self.w -= self.learning_rate * (2 * self.lambda_param * self.w)
                else:
                    # Gradient for misclassified points
                    self.w -= self.learning_rate * (2 * self.lambda_param * self.w - np.dot(x_i, y_[idx]))
                    self.b -= self.learning_rate * y_[idx]

    def predict(self, X):
        approx = np.dot(X, self.w) + self.b
        return np.sign(approx)

# Example usage
if __name__ == "__main__":
    # Define a simple dataset
    X = np.array([[1, 2], [2, 3], [3, 3], [2, 1], [3, 2]])
    y = np.array([1, 1, 1, -1, -1])

    # Train the SVM
    clf = SVM(learning_rate=0.001, lambda_param=0.01, n_iters=1000)
    clf.fit(X, y)

    # Predict on the same dataset
    predictions = clf.predict(X)
    print("Predictions:", predictions)
    print("Weights:", clf.w)
    print("Bias:", clf.b)
```
