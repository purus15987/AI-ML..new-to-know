## What is Reinforcement Learning?

* Reinforcement Learning (RL) is a type of machine learning where an **agent** learns how to make decisions by interacting with an **environment** to achieve a specific **goal**.

* The agent’s mission is to find a policy that maximizes the expected cumulative reward—often called the return—over time.

**How RL Works:**

* The **agent** observes the **state** of the **environment**.

* It takes an **action** based on a **policy**.
* The environment provides a **reward** and a **new state**.
* The agent learns by adjusting its policy to **maximize cumulative rewards**.

# Markov Decision Processes (MDP)

MDPs provide the **mathematical foundation** for Reinforcement Learning (RL).

## What is a Markov Decision Process?

Markov decision process, also called a ***stochastic dynamic program*** or ***stochastic control problem***, is a model for ***sequential decision making*** when ***outcomes are uncertain***.

* Many real-world problems involve an agent making sequential decisions in an uncertain environment to maximize rewards.

* Uncertainty: The environment may not always behave predictably.

* Goal is long-term reward: The best action is not always the one with the highest immediate reward.

At each step:

* The environment is in a specific state.
* The agent selects an action.
* The environment responds by providing a reward and transitioning to a new state, based on probabilities.

## Markov Property

A process satisfies the Markov Property ==> if the probability of reaching a new state only depends on the current state and action, not the past history.

        P(s′∣s,a) = P(s′|s,a,st−1, at−1,...)

This ensures that the system's future state only depends on the present state, making RL calculations manageable.

## Components of an MDP

An MDP is defined as a 5-tuple:

    (S,A,P,R,γ)

    S (States): The set of all possible states in the environment.

    A (Actions): The set of all possible actions an agent can take.

    P(s′|s,a) (Transition Probability): The probability of moving from state s to state s′  after taking action a.

    R(s,a) (Reward Function): The reward received after taking action a in state s.

    γ (Discount Factor): A factor between 0 and 1 that determines the importance of future rewards.

### Objective of an MDP

Find a policy (π)—a strategy that tells the agent which action to take in each state—to maximize the expected cumulative reward (also called the return). The return is the sum of all future rewards, discounted by γ:

    Return = R0 + γ.R1 + γ2.R2 +⋯


A policy can be ***deterministic*** (e.g., always move right in state s) or ***stochastic*** (e.g., move right with 70% probability).

The optimal policy (π)* is the one that achieves the ***highest possible expected return*** from any starting state.

##Understanding How an MDP Works

The agent interacts with the environment in the following cycle:

1. The agent observes the current state st.
2. It selects an action at.
3. The environment transitions the agent to a new state st+1 based on P(s′∣s,a).
4. The agent receives a reward R(s,a,s′).
5. This process repeats until the agent reaches a terminal state (like winning or losing the game).

## Solving MDPs

### Robot in a Grid World

before actual solving, let's define an "robot in a grid world" problem and solve it using different methods

**Robot in a Grid World**

The "Robot on a Grid" problem is a wellknown problem in computer science and robotics, often used to illustrate concepts in dynamic programming and graph traversal algorithms like Depth-First Search (DFS) or Breadth-First Search (BFS).

image.png

Problem Statement:
* A robot is located at the top-left corner of a m×n grid.
* The robot can only move either down or right at any point in time.
* The grid may have some cells blocked, and the robot cannot pass through them (represented as 1 in a grid).
* The task is to find a path for the robot from the top-left corner to the bottomright corner of the grid.

```python
grid = [
    [0, 0, 0],
    [0, 1, 0],  # here, (1,1) is wall
    [0, 0, 0]
]
```

1. find the shortest path from the start to the destination

Approach 1: Using Breadth-First Search (BFS)

we want to **find the shortest path from the start to the destination** in an unweighted grid.

```python
from collections import deque

def min_steps_to_reach_end(grid):
    if not grid or grid[0][0] == 1:
        return -1  # No valid path

    m, n = len(grid), len(grid[0])
    directions = [(0, 1), (1, 0)]  # Right and Down moves only
    queue = deque([(0, 0, 0)])  # (row, col, steps)
    visited = set((0, 0))

    while queue:
        r, c, steps = queue.popleft()

        if r == m - 1 and c == n - 1:  # Reached destination
            return steps

        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            if 0 <= nr < m and 0 <= nc < n and grid[nr][nc] == 0 and (nr, nc) not in visited:
                queue.append((nr, nc, steps + 1))
                print((nr, nc), steps + 1,'--->')
                visited.add((nr, nc))

    return -1  # No path found

print(min_steps_to_reach_end(grid))
```

2. count how many ways the robot can reach the destination

Approach 2: Counting Paths with Dynamic Programming

the goal is to **count how many ways the robot can reach the destination**.

DP Formula

If dp[i][j] represents the number of ways to reach (i, j), then:

    dp[i][j]=dp[i−1][j]+dp[i][j−1]

```python
def count_paths(grid):
    m, n = len(grid), len(grid[0])
    dp = [[0] * n for _ in range(m)]

    # Initialize start point
    if grid[0][0] == 1:
        return 0  # No valid path

    dp[0][0] = 1

    # Fill DP table
    for i in range(m):
        for j in range(n):
            if grid[i][j] == 1:  # Obstacle
                dp[i][j] = 0
            elif i > 0 and j > 0:
                dp[i][j] = dp[i-1][j] + dp[i][j-1]
            elif i > 0:
                dp[i][j] = dp[i-1][j]
            elif j > 0:
                dp[i][j] = dp[i][j-1]

    print(dp)

    return dp[m-1][n-1]

print(count_paths(grid))
```

3. Finding All Paths

Approach 3: Using Backtracking

If we need all possible paths, we must store the current path and backtrack when needed.

```python
def find_all_paths(grid, r, c, path, result):
    m, n = len(grid), len(grid[0])

    # If out of bounds or obstacle, return
    if r < 0 or c < 0 or r >= m or c >= n or grid[r][c] == 1:
        return

    # Add the current cell to path
    path.append((r, c))

    # If destination is reached, store the path
    if r == m - 1 and c == n - 1:
        result.append(list(path))  # Store a copy of the path
    else:
        # Move right and down
        find_all_paths(grid, r + 1, c, path, result)
        find_all_paths(grid, r, c + 1, path, result)

    # Backtrack (remove the last step)
    path.pop()

def get_all_paths(grid):
    result = []
    find_all_paths(grid, 0, 0, [], result)
    return result

paths = get_all_paths(grid)
for path in paths:
    print(path)
```

4. optimized shortest path finding

Approach 4: A* (A-star) algorithm

finds the shortest path from the start to the goal using a heuristic function (Manhattan Distance Heuristic).

    h(n) = ∣x_goal − x_current ∣ + ∣y_goal − y_current∣

```python
import heapq

def a_star(grid, start, goal):
    rows, cols = len(grid), len(grid[0])

    def heuristic(a, b):
        """Manhattan Distance"""
        return abs(a[0] - b[0]) + abs(a[1] - b[1])

    # Priority queue: (f-score, g-score, (row, col), path)
    open_list = [(0 + heuristic(start, goal), 0, start, [start])]
    heapq.heapify(open_list)
    visited = set()

    while open_list:
        f, g, (r, c), path = heapq.heappop(open_list)

        if (r, c) == goal:
            return path  # Return the shortest path

        if (r, c) in visited:
            continue
        visited.add((r, c))

        # Possible moves (Right, Down, Left, Up)
        directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]
        for dr, dc in directions:
            nr, nc = r + dr, c + dc
            if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 0:
                new_g = g + 1
                heapq.heappush(open_list, (new_g + heuristic((nr, nc), goal), new_g, (nr, nc), path + [(nr, nc)]))

    return "No path found."

# Example Grid (0 = free, 1 = obstacle)
grid = [
    [0, 0, 0, 0],
    [1, 1, 0, 1],
    [0, 0, 0, 0],
    [0, 1, 1, 0]
]

start = (0, 0)
goal = (3, 3)

# Find shortest path using A*
path = a_star(grid, start, goal)
print(path)
```

Reinforcement Learning is used when the environment is unknown and the agent must learn the best actions through experience.

### MDP Problem Statement

**MDP Problem Statement**
* The robot starts at position (0,0), and the goal is at (m-1, n-1).
* The robot can move up, down, left, or right.
* The environment has obstacles that the robot must avoid.
* The robot receives rewards for reaching the goal and penalties for hitting obstacles or invalid moves.
* The objective is to train the robot to learn the optimal policy (sequence of moves) that maximizes cumulative rewards.

```python
# Define the grid world constants
m, n = 4, 4
def print_grid():
    for i in range(m):
        row = " | ".join([f"({i},{j})" for j in range(n)])
        print(row)
        if i < m - 1:
            print("-" * (m * n))
print_grid()
```

**MDP Components**
1. **States (S):** All possible positions in the grid (e.g., (0,0), (1,2), etc.).
2. **Actions (A):** {Up, Down, Left, Right}
3. **Transition Probability (P):**
  * The environment is deterministic; the robot moves exactly as intended unless it hits a wall.
  * ELSE
    * P(s' | s, Right) = 0.8 (80% chance the robot moves right)
    * P(s' | s, Up) = 0.1 (10% chance it moves up instead)
    * P(s' | s, Down) = 0.1 (10% chance it moves down instead)
4. **Reward Function (R):**
  * +10 for reaching goal
  * -5 for hitting obstacles
  * -1 for each step
5. **Discount Factor (γ):** 0.9 (future rewards are slightly less valuable than immediate rewards).

**Goal:**

Find an optimal policy π(s) that maximizes cumulative rewards

```python
import numpy as np

# Define grid world parameters
grid_size = (5, 5)  # 5x5 grid
gamma = 0.9  # Discount factor
theta = 1e-4  # Convergence threshold

# Define rewards
goal_state = (3, 3)
obstacles = [(1, 0), (2, 2),(3,4)]  # Example obstacles
rewards = np.full(grid_size, -1)  # Default step penalty
rewards[goal_state] = 10  # Goal reward
for obs in obstacles:
    rewards[obs] = -5  # Obstacle penalty

# Define actions and stochastic transition probabilities
actions = {'U': (-1, 0), 'D': (1, 0), 'L': (0, -1), 'R': (0, 1)}
action_prob = {'R': 0.6, 'U': 0.1, 'D': 0.2, 'L': 0.1}
```

**Problem:** A robot in a grid world trying to reach the goal while minimizing steps.

**MDP Formulation:** Defined states, actions, transition probabilities, and rewards.

**Solution Methods:**

1. Dynamic Programming (DP) (requires a known model of the environment)
  * Policy Evaluation
  * Policy Iteration
  * Value Iteration
2. Monte Carlo Methods (learning from complete episodes)
3. Temporal Difference (TD) Learning (learning from incomplete episodes)

let's focus on Dynamic Programming (DP) methods for now.

### Policy Evaluation (Vπ(s))

Given a fixed policy π, we compute the **state-value function** Vπ(s) iteratively using the Bellman expectation equation:

```python
# Initialize value function
V = np.zeros(grid_size)

def get_next_state(state, action):
    """Returns the next state and reward after taking an action."""
    x, y = state
    dx, dy = action
    nx, ny = x + dx, y + dy

    # Ensure state remains within grid boundaries
    if 0 <= nx < grid_size[0] and 0 <= ny < grid_size[1]:
        return (nx, ny), rewards[nx, ny]
    return state, rewards[x, y]  # Stay in place if out of bounds

# Policy Evaluation
while True:
    delta = 0
    new_V = np.copy(V)

    for x in range(grid_size[0]):
        for y in range(grid_size[1]):
            state = (x, y)

            # Skip goal state since it's terminal
            if state == goal_state:
                continue

            v = 0
            for action, main_prob in action_prob.items():
                next_state, reward = get_next_state(state, actions[action])
                v += main_prob * (reward + gamma * V[next_state])

            new_V[x, y] = v
            delta = max(delta, abs(V[x, y] - v))

    V = new_V
    if delta < theta:
        break  # Converged

print("Value Function after Policy Evaluation:")
print(V)
```

### Policy Iteration

Finding Optimal π∗

    Step 1: Start with a random policy
    Step 2: Evaluate the policy using Bellman’s equation
    Step 3: Improve the policy by choosing better actions
    Step 4: Repeat until convergence

```python
action_list = list(actions.keys())  # List of action names
num_actions = len(actions)

# Initialize policy: equal probability for all actions
policy = np.full((grid_size[0], grid_size[1], num_actions), 1.0 / num_actions)
V = np.zeros(grid_size)

def get_next_state(state, action):
    """Returns the next state and reward after taking an action."""
    x, y = state
    dx, dy = actions[action]
    nx, ny = x + dx, y + dy

    # Ensure state remains within grid boundaries
    if 0 <= nx < grid_size[0] and 0 <= ny < grid_size[1]:
        return (nx, ny), rewards[nx, ny]
    return state, rewards[x, y]  # Stay in place if out of bounds

while True:
    # **Policy Evaluation**
    while True:
        delta = 0
        new_V = np.copy(V)

        for x in range(grid_size[0]):
            for y in range(grid_size[1]):
                state = (x, y)

                # Skip goal state (terminal state)
                if state == goal_state:
                    continue

                v = 0
                for i, action in enumerate(action_list):
                    next_state, reward = get_next_state(state, action)
                    v += policy[x, y, i] * (reward + gamma * V[next_state])

                new_V[x, y] = v
                delta = max(delta, abs(V[x, y] - v))

        V = new_V
        if delta < theta:
            break  # Converged

    # **Policy Improvement**
    policy_stable = True
    new_policy = np.copy(policy)

    for x in range(grid_size[0]):
        for y in range(grid_size[1]):
            state = (x, y)

            # Skip goal state
            if state == goal_state:
                continue

            # Compute action values
            action_values = []
            for action in action_list:
                next_state, reward = get_next_state(state, action)
                action_values.append(reward + gamma * V[next_state])

            # Select best action
            best_action = np.argmax(action_values)

            # Update policy to deterministic (all probability to best action)
            new_action_probs = np.zeros(num_actions)
            new_action_probs[best_action] = 1.0

            if not np.array_equal(policy[x, y], new_action_probs):
                policy_stable = False

            new_policy[x, y] = new_action_probs

    policy = new_policy

    if policy_stable:
        break  # Policy converged

print("\nOptimal Value Function:")
print(V)

print("\nOptimal Policy (Action Indices):")
optimal_policy = np.argmax(policy, axis=2)
print(optimal_policy)
```

### Value Iteration

Finding Optimal V*(s)

Instead of evaluating a policy, iteratively update V(s) using Bellman’s optimality equation.

```python


# Initialize value function
V = np.zeros(grid_size)

def get_next_state(state, action):
    """Returns the next state and reward after taking an action."""
    x, y = state
    dx, dy = actions[action]
    nx, ny = x + dx, y + dy

    # Ensure state remains within grid boundaries
    if 0 <= nx < grid_size[0] and 0 <= ny < grid_size[1]:
        return (nx, ny), rewards[nx, ny]
    return state, rewards[x, y]  # Stay in place if out of bounds

# **Value Iteration**
while True:
    delta = 0
    new_V = np.copy(V)

    for x in range(grid_size[0]):
        for y in range(grid_size[1]):
            state = (x, y)

            # Skip goal state (terminal state)
            if state == goal_state:
                continue

            # Compute action values
            action_values = []
            for action in action_list:
                next_state, reward = get_next_state(state, action)
                action_values.append(reward + gamma * V[next_state])

            # Update value function using the best action
            new_V[x, y] = max(action_values)
            delta = max(delta, abs(V[x, y] - new_V[x, y]))

    V = new_V
    if delta < theta:
        break  # Converged

# **Extract Optimal Policy**
policy = np.full((grid_size[0], grid_size[1]), ' ', dtype='<U1')

for x in range(grid_size[0]):
    for y in range(grid_size[1]):
        state = (x, y)

        # Skip goal state
        if state == goal_state:
            policy[x, y] = 'G'
            continue

        # Compute action values
        action_values = []
        for action in action_list:
            next_state, reward = get_next_state(state, action)
            action_values.append(reward + gamma * V[next_state])

        # Select best action
        best_action = np.argmax(action_values)
        policy[x, y] = action_list[best_action]  # Store action symbol

print("\nOptimal Value Function:")
print(V)

print("\nOptimal Policy:")
print(policy)
```

# Monte Carlo Methods

Monte Carlo methods are a class of algorithms used in model-free reinforcement learning, where the agent learns by sampling complete episodes and using the results to improve decision-making.

Unlike dynamic programming, MC methods do not require a full knowledge of the environment’s dynamics.

## Episodic Tasks

Monte Carlo methods are designed for episodic tasks, where the agent's interaction with the environment is divided into **episodes**.

Each episode starts in an initial state, proceeds through a sequence of actions and states, and ends in a terminal state, after which the environment resets.
