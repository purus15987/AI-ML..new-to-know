```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters of the Gaussian
mu = np.array([0, 0])  # Mean (center)
Sigma = np.array([[2, 1], [1, 2]])  # Covariance matrix

# Grid of (x, y) points
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# Compute the Gaussian PDF
def multivariate_gaussian(pos, mu, Sigma):
    n = mu.shape[0]
    Sigma_det = np.linalg.det(Sigma)
    Sigma_inv = np.linalg.inv(Sigma)
    norm_const = 1 / (2 * np.pi * np.sqrt(Sigma_det))
    diff = pos - mu
    return norm_const * np.exp(-0.5 * np.einsum('...k,kl,...l->...', diff, Sigma_inv, diff))

Z = multivariate_gaussian(pos, mu, Sigma)

# Plotting
fig, ax = plt.subplots(figsize=(8, 6))
contour = ax.contour(X, Y, Z, levels=5, colors='black', linewidths=1.5)
ax.contourf(X, Y, Z, levels=50, cmap="viridis")
ax.set_title("Contours and Surface of a Bivariate Gaussian", fontsize=14)
ax.set_xlabel("X-axis", fontsize=12)
ax.set_ylabel("Y-axis", fontsize=12)
ax.grid(True)
plt.colorbar(ax.contourf(X, Y, Z, levels=50, cmap="viridis"), label="PDF Value")
plt.show()
```

```python
# Plot the 3D surface of the Gaussian distribution
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Surface plot
ax.plot_surface(X, Y, Z, cmap="viridis", edgecolor='k', alpha=0.8)

# Add labels and title
ax.set_title("3D Surface of a Bivariate Gaussian Distribution", fontsize=14)
ax.set_xlabel("X-axis", fontsize=12)
ax.set_ylabel("Y-axis", fontsize=12)
ax.set_zlabel("PDF Value", fontsize=12)

plt.show()
```
